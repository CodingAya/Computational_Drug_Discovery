It is compiled from more than 88,000 documents, 1.6 million assays and the data spans 15,000 targets and 2,000 cells and 45,000 indications. [Data as of August 3, 2023; ChEMBL version 33].

```python
from chembl_webresource_client.new_client import new_client
```

# Search for Target protein

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

# Select bioactivity data for SARS coronavirus 3C-like proteinase 

```python
selected_target = targets.target_chembl_id[6]
selected_target     
```

```python
#Retrieve bioactivity data for CHEMBL3927 that are reported as IC50 values in nM (nanomolar) unit
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df.head()
```

```python
df.standard_type.unique()
```

```python
#We will save the resulting data to a CSV file
df.to_csv('bioactivity_data.csv', index=False)
```

# Handling missing data

```python
#We drop any compounds that have a missing value for the **standard_value** column
df2 = df[df.standard_value.notna()]
df2
```

# Data pre-processing

We will label compounds based on the value of their "standard_value":
- Compounds having values of less than 1000 nM will be considered to be "active" 
- Compounds having values in between 1,000 and 10,000 nM will be referred to as "intermediate"
- Compounds having values greater than 10,000 nM will be considered to be "inactive" 

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

We will now iterate the *molecule_chembl_id*, *canonical_smiles* and *standard_value* each to a list:

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

Now we are going to combine the 4 lists into a dataframe:

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
df3
```

```python
#Save preprocessed data in csv file
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
